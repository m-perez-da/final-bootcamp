```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
elec = pd.read_csv('cleaned/electricity.csv')
```

```python
elec
```

```python
futures = pd.read_csv('cleaned/futures_by_country.csv')
```

```python
futures
```

```python
steel_coal = pd.read_csv('cleaned/steel_coal_by_country.csv')
```

```python
steel_coal
```

```python
steel_coal=steel_coal.drop(['time', 'country'],axis=1)
```

```python
indicators=pd.concat([futures, steel_coal], axis=1)
```

```python
indicators['elec_price']=elec['value']
indicators['elec_var']=elec['value_var']
indicators['elec_100']=elec['value_100']
```

```python
indicators
```

```python
indicators.to_csv('cleaned/indicators_all.csv',index=False)
```

```python
indicators.to_excel('cleaned/indicators_all.xls', index=False)
```
